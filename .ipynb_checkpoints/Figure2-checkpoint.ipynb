```python
import numpy as np
np.random.seed(1990)
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('./data/processed/cell_type_clustering.csv', index_col=0)
```

```python
df.head()
```

```python
# do clustering, check preplot

all_data = df[['diff', 'soma_size', 'field','asym']].values
X = all_data.copy()

X = (X - X.min(0)) / (X.max(0) - X.min(0))

l0 = KMeans(n_clusters=2).fit(X[:, 0].reshape(-1,1))

# tOffa vs mini a
X1 = df[l0.labels_ == 0][['diff', 'soma_size', 'field','asym']].values
idx1 = df[l0.labels_ == 0].index
l1 = KMeans(n_clusters=2).fit(X1[:, 1].reshape(-1,1))
# l1 = KMeans(n_clusters=2).fit(X1[:, 1:3])

# fmini vs sustain
X2 = df[l0.labels_ == 1][['diff', 'soma_size', 'field','asym']].values
idx2 = df[l0.labels_ == 1].index
l2 = KMeans(n_clusters=2).fit(X2[:, 3].reshape(-1,1))

g0 = df.loc[idx1[l1.labels_==0]]
g1 = df.loc[idx1[l1.labels_==1]]
g2 = df.loc[idx2[l2.labels_==0]]
g3 = df.loc[idx2[l2.labels_==1]]

fig = plt.figure(figsize=(8.27,8.27))
ax = plt.subplot(111)

celltypes = ['tOff-alpha','mini-alpha','f-mini', 'sOff-alpha' ]

for i, g in enumerate([g0, g1, g2, g3]):
    expdate = g[['date']].values
    x = g[['field']].values
    y = g[['diff']].values

    z = g[['soma_size']].values
    z /= df[['soma_size']].values.max()
    
    ax.scatter(x,y,s=z*25**2, label=celltypes[i],  edgecolors='white')
    
    for j, txt in enumerate(expdate):
#         print(txt)
        ax.annotate(txt[0], xy=(x[j], y[j]),xytext=(x[j]+0.2, y[j]), fontsize=9)
ax.legend(frameon=False)

    
ax.set_ylabel('Marginal Arbor Density - Central Arbor Density')
ax.set_xlabel('Dendritic Field Size  x 1000 (um^2)')


ax.set_ylabel('Marginal Arbor Density - Central Arbor Density')
ax.set_xlabel('Dendritic Field Size  x 1000 (um^2)')

ax.set_ylabel('Marginal Arbor Density - Central Arbor Density')
ax.set_xlabel('Dendritic Field Size  x 1000 (um^2)')

ax.set_xlim(5, 80)
ax.set_ylim(-15, 15)
    
ax.set_ylabel('Marginal Arbor Density - Central Arbor Density')
ax.set_xlabel('Dendritic Field Size  x 1000 (um^2)')

ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(0)
ax.spines['top'].set_linewidth(0)
ax.tick_params(labelsize=18, width=2, length=10)
```
