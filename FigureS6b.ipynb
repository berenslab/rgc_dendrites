```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

filepath = './data/processed/overlap.csv'

label_fontsize = 22
tick_fontsize = 18
cbar_fontsize = 16

df = pd.read_csv(filepath, index_col=0)

vmin = -400
vmax = 800
```

# toffa

```python
toffa = df[df.type == 0]
x = toffa['dendritic_distance_between_rois'].values
y = toffa['cbpt_angle_between_rois_deg'].values
z = toffa['overlap_index'].values

xlim_fixed = int(np.ceil(x.max() / 100.0)) * 100
ylim_fixed = int(np.ceil(y.max() / 100.0)) * 100

ratiox = round(16 * x.max() / xlim_fixed).astype(int)
ratioy = round(10 * y.max() / ylim_fixed).astype(int)
gSize=[ratiox, ratioy]

fig, ax = plt.subplots(1,1, figsize=(8,8))


im1 = ax.hexbin(x, y, gridsize=gSize, mincnt=1, vmin=vmin, vmax=vmax, cmap=plt.cm.Greys)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.tick_params(labelsize=tick_fontsize, width=2)

ax.set_xlabel('Dendritic distance between ROIs', fontsize=label_fontsize*0.8)
ax.set_ylabel('Angle between ROIs', fontsize=label_fontsize*0.8)

ax.set_xlim(-30,xlim_fixed)
ax.set_ylim(-10,ylim_fixed)
sns.despine(offset=10, trim=True)
```
