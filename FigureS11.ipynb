```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
```

# toffa

```python
data_toffa = pickle.load(open('./data/model/toffa.pickle', 'rb'))
for ii, key0 in enumerate(data_toffa.keys()):
    
    fig, ax = plt.subplots(3,2, figsize=(6,8))
    ax = ax.flatten()    
    
    for jj, key1 in enumerate(data_toffa[key0].keys()):
        
        d = data_toffa[key0][key1]
        ax[jj].imshow(d, origin='lower', vmin=0.0, vmax=1)
        ax[jj].set_title(key1)
    
    fig.suptitle(key0)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    
```
