```python
import numpy as np
np.random.seed(1990)
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from morphkit import Morph
from matplotlib_scalebar.scalebar import ScaleBar

from sklearn.cluster import KMeans
df = pd.read_csv('./data/processed/cell_type_clustering.csv', index_col=0)
```

```python
np.unique(df.clustering_result)
```

```python
toffa = df[df['clustering_result'] == 'tOFFa']['date'].values
soff = df[df['clustering_result'] == 'sOff']['date'].values
minia = df[df['clustering_result'] == 'mini-alpha']['date'].values
fmini = df[df['clustering_result'] == 'f-mini']['date'].values
```

# tOffa

```python
fig, ax = plt.subplots(5, 4, figsize=((8.27, 11.69))) 
ax = ax.flatten()
for i, exp in enumerate(toffa):
    path = './data/morphologies/' + '_'.join(exp.split('/')) + '.swc'
    m = Morph(path)
    soma_pos = m.df_paths[m.df_paths.type == 1]['path'][0][0]
    dendrites = m.df_paths[m.df_paths.type == 3]
    for row in dendrites.iterrows():

        path_id = row[0]

        path = row[1]['path']
        ax[i].plot(path[:, 0], path[:, 1], color='black', lw=1)
    ax[i].scatter(soma_pos[0], soma_pos[1], c='blue', s=60, zorder=10)    
    ax[i].set_title('{}'.format(exp))
    
    ax[i].set_xlim(-50, 350)
    ax[i].set_ylim(-50, 350)
    ax[i].set_aspect('equal', 'box')
    
    scalebar = ScaleBar(1, units='um', location='lower left', box_alpha=0, pad=0, length_fraction=0.15)
    ax[i].add_artist(scalebar)
    ax[i].invert_yaxis()
    
    ax[i].axis('off')

for redundent in range(i+1, 20):
    ax[redundent].set_aspect('equal', 'box')
    ax[redundent].axis('off')
```
