```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import h5py
```

```python
label_fontsize = 22
tick_fontsize = 18
cbar_fontsize = 16

cmap = plt.cm.jet
vmin=-1
vmax=1
```

# lchirp

## tOffa

```python
filepath = './data/processed/lchirp_corr.csv'

df = pd.read_csv(filepath, index_col=0)

toffa = df[df.type == 0]
x = toffa['lChirpCorr_distance'].values
y = toffa['lChirpCorr_Angle'].values
z = toffa['lChirp_corr'].values

xlim_fixed = int(np.ceil(x.max() / 100.0)) * 100
ylim_fixed = int(np.ceil(y.max() / 100.0)) * 100

ratiox = round(16 * x.max() / xlim_fixed).astype(int)
ratioy = round(10 * y.max() / ylim_fixed).astype(int)
gSize=[ratiox, ratioy]

fig, ax = plt.subplots(1,1, figsize=(8,8))
im1 = ax.hexbin(x, y, C=z, gridsize=gSize, alpha=0.8, vmin=vmin, vmax=vmax, cmap=cmap)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.tick_params(labelsize=tick_fontsize, width=2)

ax.set_xlabel('Dendritic distance between ROIs', fontsize=label_fontsize*0.8)
ax.set_ylabel('Angle between ROIs', fontsize=label_fontsize*0.8)

ax.set_xlim(-30,xlim_fixed)
ax.set_ylim(-10,ylim_fixed)
sns.despine(offset=10, trim=True)
```
