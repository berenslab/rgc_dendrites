```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
filepath = './data/processed/offset_and_angle.csv'
label_fontsize = 22
tick_fontsize = 18
cbar_fontsize = 16

df = pd.read_csv(filepath, index_col=0)
df = df.drop(1911)
vmin = 0
# vmax = df['offset'].max()
vmax = 180
```

# toffa

```python
toffa = df[df.cell_type == 0]
x = toffa['dendritic_distance_to_soma'].values
z = toffa['angles_btw_RFcenter_and_DensityCenter'].values
y = toffa['offset'].values

# xlim_fixed = int(np.ceil(x.max() / 100.0)) * 100
xlim_fixed = 310
# ylim_fixed = int(np.ceil(y.max() / 100.0)) * 100
ylim_fixed = 175

ratiox = round(16 * x.max() / xlim_fixed).astype(int)
ratioy = round(10 * y.max() / ylim_fixed).astype(int)
gSize=[ratiox, ratioy]

fig, ax = plt.subplots(1,1, figsize=(8,8))

# from matplotlib.gridspec import GridSpec
# fig = plt.figure(figsize=(8,8))

# gs = GridSpec(4,4)
# ax = fig.add_subplot(gs[1:4,0:3])

im1 = ax.hexbin(x, y, C=z, gridsize=gSize, alpha=0.8, vmin=vmin, vmax=vmax)
# ax.scatter(x, y, c=z, vmin=0, vmax=1, s=5)

# cb = fig.colorbar(im1, ax=ax, drawedges=False)
# cb.set_label("Overlap Index", fontsize=cbar_fontsize*0.8)
# cb.outline.set_visible(False)
# cb.solids.set_edgecolor("face")

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.tick_params(labelsize=tick_fontsize, width=2)

ax.set_xlabel('Dendritic distance to soma', fontsize=label_fontsize*0.8)
ax.set_ylabel('Offset', fontsize=label_fontsize*0.8)

ax.set_xlim(-30,xlim_fixed)
ax.set_ylim(-10,ylim_fixed)
sns.despine(offset=10, trim=True)
```

```python
acute = toffa[toffa['angles_btw_RFcenter_and_DensityCenter'] < 90]
obtuse = toffa[toffa['angles_btw_RFcenter_and_DensityCenter'] >= 90]
hist_acute, bins, _ = plt.hist(acute['dendritic_distance_to_soma'], bins=30, facecolor='none', edgecolor='black')
_, _, _ = plt.hist(obtuse['dendritic_distance_to_soma'], bins=bins, facecolor='none', edgecolor='red')
plt.ylabel('# Angles')
plt.xlabel('Dendritic distance from ROI to soma')
plt.xlim(0, 300)
plt.ylim(0, 60)
```
